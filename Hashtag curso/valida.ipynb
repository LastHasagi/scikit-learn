```python
import pandas as pd

# Carrega a base de dados
tabela = pd.read_excel("tabela.xlsx")

# Verifica se há valores vazios ou valores reconhecidos em formato errado
colunas_excluir = ['Unnamed: 0', 'Matriz ou Filial', 'EXCEÇÃO', 'Razao Social', 'Nome Fantasia', 'Data Situacao', 'Abertura', 'DDD1', 'Telefone1', 'DDD2', 'Telefone2', 'Conversa', 'contato ativo?', 'E-mail', 'Socio']
tabela = tabela.drop(columns=colunas_excluir)
print(tabela.info())
print(tabela.columns)
display(tabela)

import numpy as np
from sklearn.preprocessing import LabelEncoder

# Criando o codificador para transformar colunas de texto em números
codificador = LabelEncoder()

# Mapeamento para a coluna "Situacao"
mapeamento_situacao = {'ATIVA': 1}
tabela['Situacao'] = tabela['Situacao'].map(mapeamento_situacao)

# Mapeamento para a coluna "Valida"
mapeamento_valida = {'APROVADO': 0, 'NÃO APROVADO': 1}
tabela['Valida'] = tabela['Valida'].map(mapeamento_valida)

# Verificar informações sobre o DataFrame após as modificações
print(tabela.info())

# Escolhendo quais colunas vamos usar para treinar o modelo
# y é a coluna que queremos que o modelo calcule
# x são todas as colunas que vamos usar para prever o Capital Social
x = tabela.drop(['Capital Social', 'CNPJ'], axis=1)
y = tabela['Capital Social']

from sklearn.model_selection import train_test_split

# Separamos os dados em treino e teste. Treino vamos dar para os modelos aprenderem e teste vamos usar para ver se o modelo aprendeu corretamente
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=5)

from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.impute import SimpleImputer

# Verificar se há valores ausentes no conjunto de treinamento
if x_treino.isnull().values.any() or y_treino.isnull().values.any():
    # Aplicar o imputer aos dados de treinamento
    imputer = SimpleImputer(strategy='mean')
    x_treino_imputado = imputer.fit_transform(x_treino)
    x_treino = pd.DataFrame(x_treino_imputado, columns=x_treino.columns)

# Criar e treinar o modelo HistGradientBoostingClassifier
modelo_boosting = HistGradientBoostingClassifier()
modelo_boosting.fit(x_treino, y_treino)

# Se o nosso modelo chutasse tudo "Standard", qual seria a acurácia do modelo?
contagem_valida = tabela['Valida'].value_counts()
print(contagem_valida[0] / sum(contagem_valida))

from sklearn.metrics import accuracy_score

# Calculamos as previsões
previsao_boosting = modelo_boosting.predict(x_teste)

# Comparamos as previsões com o y_teste
# Esse score queremos o maior (maior acurácia, mas também tem que ser maior do que o chute de tudo Standard)
print(accuracy_score(y_teste, previsao_boosting))

from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Criar e treinar o modelo RandomForestClassifier
modelo_arvore = RandomForestClassifier()
modelo_arvore.fit(x_treino, y_treino)

# Criar e treinar o modelo KNeighborsClassifier
modelo_knn = KNeighborsClassifier()
modelo_knn.fit(x_treino, y_treino)
```
